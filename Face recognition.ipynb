```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.image as mpimg
from deepface import DeepFace
import cv2 as cv
```

### Testing

```python
aadhar_no = input("Enter Aadhar Number:")

c = plt.imread("2.jpg")


aadhar_image = 'photos/' + aadhar_no + '.jpg'


result = DeepFace.verify(c,aadhar_image,model_name="Facenet",
                         enforce_detection=False,distance_metric='euclidean_l2',detector_backend = 'retinaface')

if result['verified']:
    print("Same person")
    images = [c,plt.imread(aadhar_image)]
    figure, plots = plt.subplots(1,2,figsize=(10,10))
    for i in range(2):
          plots[i].imshow(images[0])
          plots[i].axis('off')
          images.pop(0)
else:
    print("Person did not match")
```

### Dynamic Checking (Video)

```python
video = cv.VideoCapture(0)
aadhar_no = input("Enter Aadhar Number:")

while True:
    ret, image = video.read()
    #c = plt.imread("sample2.jpg")

    aadhar_image = 'data/' + aadhar_no + '.jpg'

    result = DeepFace.verify(image,aadhar_image,model_name="Facenet",enforce_detection=False,
                         distance_metric='euclidean_l2',detector_backend = 'retinaface')

    print(result['verified'])
    #cv.imshow("Video",cv.cvtColor(image,cv.COLOR_BGR2RGB))


    cv.imshow("video",image)
    if cv.waitKey(1) and 0xFF == ord('q'):
        break

video.release()
cv.destroyAllWindows()
```
