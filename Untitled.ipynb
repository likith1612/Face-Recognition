```python
import cv2 as cv
import matplotlib.pyplot as plt
c = cv.VideoCapture(0)
while True:
    ret,frame = c.read()
#     face = cv.CascadeClassifier(cv.data.haarcascades+"haarcascade_frontalface_default.xml")
#     gr = cv.cvtColor(frame,cv.COLOR_RGB2GRAY)
#     faces = face.detectMultiScale(gr,scaleFactor = 1.25,minNeighbors = 2)
#     for x,y,w,h in faces:
#         face_1 = frame[y-100:y+h+100,x-100:x+w+100]
#     cv.imshow("video",face_1)
    cv.imshow("video",frame)
    if(cv.waitKey(5)==ord('l')):
#         cv.imwrite("test_1.jpg",cv.cvtColor(face_1,cv.COLOR_BGR2GRAY))
        cv.imwrite("test_1.jpg",frame)
        break
c.release()
cv.destroyAllWindows()
```

```python
verify = DeepFace.verify(img1_path="test_1.jpg",img2_path=test,model_name='Facenet',
                         enforce_detection=False,distance_metric='euclidean_l2',detector_backend = 'retinaface')
verify
```

```python
x = verify.get('verified')
img = cv.cvtColor(cv.imread("test_1.jpg"),cv.COLOR_BGR2RGB)
if(x):
    print("Person verified")
    print("Aadhar no.: "+aadhar_inp)
    plt.subplot(1,2,1)
    plt.imshow(test_img)
    plt.axis('off')
    plt.subplot(1,2,2)
    plt.imshow(img)
    plt.axis('off')
else:
    print("Person not matched")
    plt.subplot(1,2,1)
    plt.imshow(test_img)
    plt.axis('off')
    plt.subplot(1,2,2)
    plt.imshow(img)
    plt.axis('off')
```
